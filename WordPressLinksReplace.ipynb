```python
import pandas as pd
```

```python
# Read in the .csv and extract the old and new link pairs

linksReplaceDF = pd.read_csv('Squiz links audit(Links on Squiz).csv')
linksReplaceDF = linksReplaceDF[['Current Squiz link', 'New squiz link']]
linksReplaceDF = linksReplaceDF.dropna(how='any')
linksReplaceDF.reset_index(drop='True')
```

```python
# Script to see how many times these links appear. Effectively to check we are targeting the right table/fields for update.

sqlScript = 'SELECT COUNT(*) FROM wp_303_posts WHERE '


firstRow = True
 
# Generate the query.
for index, row in linksReplaceDF.iterrows():
    
    # We don't want a phantom 'OR' at the end of our query.
    if firstRow:
        firstRow = False
    else:
        sqlScript += ' OR '
    
    sqlScript += 'post_content LIKE \'%' + row['Current Squiz link'] + '%\''
    

sqlScript += ';'
```

```python
# Script to update all the links across the wp_xxx_posts table.

sqlScript = ''
siteID = '303' # So we can target different sites if we need.


# Generate the query.
for index, row in linksReplaceDF.iterrows():
    
    sqlScript += '\n\n' + 'UPDATE wp_' + siteID +'_posts SET post_content = REPLACE(post_content, \'' + row['Current Squiz link'] +  '\', \'' + row['New squiz link'] +  '\') WHERE post_content LIKE \'%' + row['Current Squiz link'] + '%\';'


print(sqlScript)
```
